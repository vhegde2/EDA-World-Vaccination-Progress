```python
covidv["people_vaccinated"]= covidv.groupby("country").total_vaccinations.tail(1)
vaccination_country=covidv.groupby("country")["people_vaccinated"].mean().sort_values(ascending= False).head(10).reset_index()
```

```python
vaccination_country
```

```python
#Top 10 countries with highest total vaccinations
covidv["Total_vaccinations(count)"]= covidv.groupby("country").total_vaccinations.tail(1)
covidv.groupby("country")["Total_vaccinations(count)"].mean().sort_values(ascending= False).head(10)
```

```python
#This is correct
x= covidv.groupby("country")["Total_vaccinations(count)"].mean().sort_values(ascending= False).head(10)
sns.set_style("whitegrid")
plt.figure(figsize= (8,8))
ax= sns.barplot(x.values,x.index)
ax.set_xlabel("Total vaccinations(count)")
plt.show()
```

```python
#Top 10 countries
x=covidv.groupby("country")["total_vaccinations"].sum().sort_values(ascending=False).head(10)
sns.set_style("whitegrid")
plt.figure(figsize= (8,8))
ax= sns.barplot(x.values,x.index)
ax.set_xlabel("Total vaccinations(count)")
plt.show()
```

```python
#Top 10 countries highest fully vaccinated 
covidv["Full_vaccinations"]= covidv.groupby("country").people_fully_vaccinated.tail(1)
covidv.groupby("country")["Full_vaccinations"].mean().sort_values(ascending= False).head(10)
x= covidv.groupby("country")["Full_vaccinations"].mean().sort_values(ascending= False).head(10)
sns.set_style("whitegrid")
plt.figure(figsize= (10,10))
ax= sns.barplot(x.values,x.index)
ax.set_xlabel("Fully vaccinated")
plt.show()
```
